### `profile`

```python
import numpy as np
import pandas as pd

pd.options.display.max_columns = None
```

```python
data = {
    "a": [True, None, False, False, True, False],
    "b": [1, 2, 3, 4, 5, 6],
    "c": [1.0, 2.0, 3.0, None, 5.0, 2.0],
    "d": pd.Categorical([1, 2, 1, None, 1, 2]),
    "e": ["alice", "bob", None, "eric", "bob", None],
    "f": [0] * 6,
    "g": [None] * 6,
    "h": [float("nan")] * 6,
    "i": [np.nan] * 6,
    "j": pd.Categorical([1, 2, 1, 1, 1, 1]),
    "k": ["2", "1", "2", "1", "2", "1"],
}
df = pd.DataFrame(data)
df
```

```python
summary_df = pandastools.profile(df)
summary_df.T
```
